```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

# Answer:-

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
size = 100
x = np.random.normal(0, 1, size)
y =np.random.normal(0, 0.5, size)  # y is correlated with x
z =np.random.normal(0, 1, size)  # z is less correlated with x

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Compute the correlation matrix
corr_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)

# Display the plot
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
# np.random.seed(20)
# data = {
#     'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#     'Day': np.random.choice(range (1, 31), 100),
#     'Sales': np.random.randint(1000, 5000, 100)
# }
# df pd.DataFrame(data)


# Answer:-


import numpy as np
import pandas as pd
import plotly.express as px

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the Data for Heatmap
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='sum')

# Create Heatmap
fig = px.imshow(heatmap_data, labels=dict(x="Month", y="Day", color="Sales"),title='Heatmap of Sales Across Different Months and Days')
fig.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(x² + y²)
# x= np.linspace (-5, 5, 100)
# y = np.linspace(-5, 5, 100)
# x, y np.meshgrid(x, y)
# znp.sin(np.sqrt(x*2 + y*2))

# data = {
#     'X': x.flatten(),
#     'Y': y.flatten(),
#     'Z': z.flatten()
# } 
# df pd. DataFrame (data)


# Answer:-



import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create 3D Surface Plot
fig = go.Figure()
fig.add_trace(go.Surface(x=x,y=y,z=z))
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()
```

BOKEH ASSIGNMENT:

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

# Answer:-

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, legend_label="sin(x)", line_width=2)
output_notebook()
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

# Answer:-

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100) * 50
colors = np.random.rand(100)
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))
p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y', width=800, height=400)
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6,source=source)

# Show the plot
output_notebook()
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits ['Apples', 'Oranges', 'Bananas', 'Pears'] 
# counts [20, 25, 30, 35]


# Answer:-

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
# Prepare the data source
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', width=800, height=400)
p.vbar(x='fruits', top='counts', width=0.5,source=source, color="navy", alpha=0.7)
output_notebook()
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist np.random.randn (1000)
# hist, edges np.histogram(data_hist, bins=30)


# Answer:-


import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot
p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

# Plot the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
output_notebook()
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
# data_heatmap np.random.rand(10, 10)
# x = np.linspace(0, 1, 10)
# y= np.linspace(0, 1, 10)
# xx, yy np.meshgrid(x, y)


# Answer:-


import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(values), high=max(values))

# Create a new plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=600, height=600)
p.rect(x='x', y='y', width=0.1, height=0.1, source=dict(x=x_flat, y=y_flat, value=values),
       fill_color=mapper, line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
output_notebook()
show(p)
```
